```
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.style.use('seaborn-whitegrid')
```

#Simple Line Plots

```
yasListesi = np.array([10, 20, 30, 40, 50, 60, 70, 80, 90])
kiloListesi = np.array([20, 60, 80, 85, 85, 65, 45, 85, 63])
plt.plot(yasListesi, kiloListesi, "r")
plt.xlabel("Yas")
plt.ylabel("Kilo")
#plt.show()
plt.title("Kilonun yaşa göre değişim grafiği")
```

```
plt.plot(kiloListesi, yasListesi, "r")
plt.title("Yaşın Kiloya göre değişim grafiği")
```

```
dizi = np.linspace(0,100,20)
dizi2 = dizi**2
plt.plot(dizi,dizi2, "g*-")
```

```
plt.subplot(1,2,1)
plt.plot(dizi,dizi2, "b--")
plt.subplot(1,2,2) # 1 sıra, 2 kolon, şu an 2. grafikteyim
plt.plot(dizi2,dizi, "g*-")
```

```
#Kendi figürümü üretebilirim
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.5,0.5,0.7,0.7])
figureAxes.plot(dizi,dizi2,"g")
figureAxes.set_xlabel("X Ekseni")
figureAxes.set_ylabel("Y Ekseni")
figureAxes.set_title("Başlık")
```

```
#Üst üste figürler
figure2 = plt.figure()
eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2 = figure2.add_axes([0.2,0.4,0.3,0.3])

eksen1.plot(dizi,dizi2,"g")
eksen1.set_xlabel("X ekseni")
eksen1.set_ylabel("Y ekseni")
eksen1.set_title("Baslik")

eksen2.plot(dizi,dizi2,"b")
eksen2.set_xlabel("X ekseni")
eksen2.set_ylabel("Y ekseni")
eksen2.set_title("Baslik")
```

```
(benimFigure, benimEksen) = plt.subplots(nrows=1, ncols=2)
for eksen in benimEksen:
  eksen.plot(dizi,dizi2,"g")
  eksen.set_xlabel("X ekseni")
plt.tight_layout()
```

```
yeniFigur = plt.figure(dpi=100)
#dpi piksel sayısıdır. Piksel arttıkça kaliteli olur
yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(dizi,dizi2,label="dizi**2")
yeniEksen.plot(dizi,dizi**3,label="dizi**3")
yeniEksen.legend()
```

```
#Figürümü yüksek kalitede kaydedebilirim
yeniFigur.savefig("benimFigur.png", dpi=200)
```

##Renkler ve Stiller

```
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

#Simple Scatter Plots

##Scatter Plots with plt.plot

```
#3. argüman kullanılan semboldür. '-', '--' gibi değer de alabilir. 
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black');
```

```
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8);
```

```
plt.plot(x, y, '-ok');
```

##Scatter Plots with plt.scatter

```
plt.scatter(x, y, marker='o');
```

```
#alfa argümanı şeffaflık için kullanılır
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
plt.colorbar();  # show color scale
```

```
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.3, normed=True, bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```
